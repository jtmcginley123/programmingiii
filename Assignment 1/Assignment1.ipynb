#### Q1.Return the first 7 records from the housing table

#### Q2. How many records in the data had sales greater than 100 and a median sales price greater than 100000? Write a query to return the result.

#### A2.

```python
%%sql

select count(*) 

from housing 

where sales > 100
and median > 100000
```

#### Q3. Show the number of records in the data that had sales greater than 100 or a median sales price less than 100000 by month. The result should contain two columns: month and num_records.

#### A3.

```python
%%sql

select 
month,
count(*) as num_records 

from housing 

where sales > 100
and median > 100000

group by month
```

#### Q4. Return a result that shows the mean annual sales by city and year, excluding cities that begin with the letter H. The result should be in descending order by mean sales, limited to 20 records, and should contain three columns: city, year, and mean_annual_sales

#### A4.

```python
%%sql

select 
    city, 
    year,
    round(avg(sales), 2) AS mean_annual_sales

from housing 

where 
    city not like 'H%'

group by 
    city, 
    year

order by 
     mean_annual_sales DESC

limit 20

```

#### Q5. How many sales were there in Austin in December 2010? Return the city, year, and sales in the result.

#### A5. 

```python
%%sql

select sales 

from housing 

where city = 'Austin'
and month = 12
and year = 2010
```

#### Q6. In 2010 in Austin, which month had the most sales? Return a result that shows the number of sales in Austin in 2010 by month.

#### A6.

```python
%%sql

select month, max(sales)

from housing

where city = 'Austin'
and year = 2010
```

#### Q7. In which city-month combinations did the greatest numbers of sales, outside of Austin? Return a result that contains the top five cities’ sales and the month it took place in 2010, excluding Austin.

#### A7.

```python
%%sql

select 
    city, 
    month, 
    sales

from housing 

where
    city <> 'Austin'
and
    year = 2010

order by sales desc

limit 5
```

#### Q8. In which city, year, and month did the fewest sales take place? The result should return four columns: fewest_sales, city, year, and month

#### A8.

```python
%%sql

select
    min(sales) as fewest_sales,
    city,
    year,
    month

from housing
```

#### Q9. For the entire dataset, what were the mean sales in each year?

#### A9.

```python
%%sql

select
    year,
    round(avg(sales), 2) as avg_sales

from housing

group by year
```

#### Q10. Return only the years that had mean sales greater than 600. Which years does this include?

#### A10.

```python
%%sql

select year

from housing

group by year

having avg(sales) > 600

order by year
```
