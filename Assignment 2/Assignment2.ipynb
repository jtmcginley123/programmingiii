#### Q16. On average, what is the most expensive item purchased? Write a query to show the mean sale price of each item in the orders table and sort the results so the most expensive item is first and the least expensive item is last.

#### A16.

```python
%%sql sqlite:///shop.db

select item_name as "Item Name", avg(price) as "Average Price"

from orders 

group by item_name

order by avg(price) desc
```

#### Q20. We need to know more about the customers who purchased the items in the orders table. Write a join to join the orders table to the customers table. The results should show only customers who have made a purchase that is listed in the orders table and are listed in the customers table and the item_name they purchased. (Note that not all customer ids are listed in the customer table.)

#### A20.

```python
%%sql sqlite:///shop.db

select C.customer_id as "Customer ID", item_name as "Item Name"

from  orders O
left join customers C on O.customer_id = C.customer_id

where  C.customer_id is not null
```

#### Q21. Alter the code you wrote in #20 to group the results by customer id from the customers table and count the number of orders for each customer made. Alias the result column as “num_orders” and sort the results so the customer with the most orders is the first row in the result. Which customer id has the most orders?

#### A21.

```python
%%sql sqlite:///shop.db

select C.customer_id as "Customer ID", count(order_id) as "num_orders"

from  orders O
left join customers C on O.customer_id = C.customer_id

where  C.customer_id is not null

group by C.customer_id

order by num_orders desc;

```

#### Q24. Write a subquery (i.e. nested query) to filter the customers table to keep only records that are less than the average age for all customers in the table

#### A24. 

```python
%%sql sqlite:///shop.db

select *

from customers

where age < (
    select avg(age)
    from customers
);
```

#### A25. 

```python
%%sql sqlite:///shop.db

select
    country as "Country",
    avg(age) AS "mean_age"

from customers

group by country
having avg(age) > 60;
```
