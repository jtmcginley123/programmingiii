#### Q4. As you build your relational database, what are three important components to consider? Describe each of these three components and why they are important.

#### A4. 

#### The three important components to consider are 1. Tables, 2. Keys, 3. Relationships/Constraints

#### 1. Tables store data in rows and columns, and they can represent a specific importance to a business. Tables are important because they provide a clear structure for storing data and make data easy to query. 

#### 2. Keys are important for differnet reasons. There are primary keys which are important for uniquely identifying each row in a table. Foreign keys are important for linking one table to another. Both of these together help maintain data integrity.

#### 3. Relationships and constrains are important because they are rules that define how tables relate to one another and what values are allowed to be considered when analyzing data. They enforce data consistency and accuracy.
